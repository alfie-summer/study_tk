```python
import tkinter as tk

window = tk.Tk()
window.title('Cool Calculator')
##窗口尺寸
window.geometry('300x400')

inputs = []

# 创建一个显示屏，用于显示过程和结果
label = tk.Label(window,
    text='Please input your math expression',    # 标签的文字
    width=260, height=2,     # 标签长宽
    borderwidth=2,
    relief="ridge"
    )
label.pack()  


def click_button_digit(x):
    if x in "01234567890+-*/.":
        inputs.append(x)
        label.config(text="".join(inputs))


def clear():
    inputs.clear()
    label.config(text="Please input your math expression")
    

def evaluate():
    try:
        r = eval("".join(inputs))
        label.config(text=str(r))
        inputs.clear()
    except:
        label.config(text="ERROR")
    

    
def backspace():
    if inputs:
        inputs.pop(-1)
    if not inputs:
        label.config(text="Please input your math expression")

button1 = tk.Button(window,
    text='1',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:click_button_digit('1'))     # 点击按钮时执行的函数
button1.pack()
button1.place(x=10, y=50)

button2 = tk.Button(window,
    text='2',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:click_button_digit('2'))     # 点击按钮时执行的函数
button2.pack()
button2.place(x=60, y=50)

button3 = tk.Button(window,
    text='3',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:click_button_digit('3'))     # 点击按钮时执行的函数
button3.pack()
button3.place(x=110, y=50)


button4 = tk.Button(window,
    text='4',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:click_button_digit('4'))     # 点击按钮时执行的函数
button4.pack()
button4.place(x=10, y=110)

button5 = tk.Button(window,
    text='5',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:click_button_digit('5'))     # 点击按钮时执行的函数
button5.pack()
button5.place(x=60, y=110)

button6 = tk.Button(window,
    text='6',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:click_button_digit('6'))     # 点击按钮时执行的函数
button6.pack()
button6.place(x=110, y=110)

button7 = tk.Button(window,
    text='7',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:click_button_digit('7'))     # 点击按钮时执行的函数
button7.pack()
button7.place(x=10, y=170)

button8 = tk.Button(window,
    text='8',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:click_button_digit('8'))     # 点击按钮时执行的函数
button8.pack()
button8.place(x=60, y=170)

button9 = tk.Button(window,
    text='9',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:click_button_digit('9'))     # 点击按钮时执行的函数
button9.pack()
button9.place(x=110, y=170)

buttonsign = tk.Button(window,
    text='+/-',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:click_button_digit('+/-'))     # 点击按钮时执行的函数
buttonsign.pack()
buttonsign.place(x=10, y=230)

button0 = tk.Button(window,
    text='0',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:click_button_digit('0'))     # 点击按钮时执行的函数
button0.pack()
button0.place(x=60, y=230)

buttonpoint = tk.Button(window,
    text='.',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:click_button_digit('.'))     # 点击按钮时执行的函数
buttonpoint.pack()
buttonpoint.place(x=110, y=230)


buttonplus = tk.Button(window,
    text='+',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:click_button_digit('+'))     # 点击按钮时执行的函数
buttonplus.pack()
buttonplus.place(x=160, y=50)

buttonminus = tk.Button(window,
    text='-',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:click_button_digit('-'))     # 点击按钮时执行的函数
buttonminus.pack()
buttonminus.place(x=160, y=110)

buttonmul = tk.Button(window,
    text='*',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:click_button_digit('*'))     # 点击按钮时执行的函数
buttonmul.pack()
buttonmul.place(x=160, y=170)

buttondiv = tk.Button(window,
    text='/',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:click_button_digit('/'))     # 点击按钮时执行的函数
buttondiv.pack()
buttondiv.place(x=160, y=230)

buttonplus = tk.Button(window,
    text='C',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:clear())     # 点击按钮时执行的函数
buttonplus.pack()
buttonplus.place(x=210, y=50)

buttonmplus = tk.Button(window,
    text='M+',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:set_memory())     # 点击按钮时执行的函数
buttonmplus.pack()
buttonmplus.place(x=210, y=110)

buttonmr = tk.Button(window,
    text='MR',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:restore_memory())     # 点击按钮时执行的函数
buttonmr.pack()
buttonmr.place(x=210, y=170)

buttonequ = tk.Button(window,
    text='=',             # 显示在按钮上的文字
    width=5, height=2,
    command=lambda:evaluate())     # 点击按钮时执行的函数
buttonequ.pack()
buttonequ.place(x=210, y=230)

##显示出来
window.mainloop()
```
